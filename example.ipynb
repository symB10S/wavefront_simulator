```python
import wavefronts.generation as gen
import wavefronts.plotting as wp
import wavefronts.interactive as wi
import matplotlib.pyplot as plt

# this is essential to make plots interactive
%matplotlib widget
```

Generate a resonator circuit's distibuted response:

```python
interface = gen.generate_interface_data(L_impedance = '985',L_time='0.3',C_impedance='0.88',C_time='0.53')
```

interact with the spatial disribution across the capacitor and inductor:

```python
wi.interact_spatial(interface)
```

Interact with fanout diagrams:

```python
wi.interact_fanout_path(interface,False)
```

inspect reflection diagrams:

```python
fig_ref,ax_ref = plt.subplots()
wp.plot_refelction_diagram(interface,ax_ref,False)
```

How about an interesting 3D plot? 

```python
wi.interact_3D_spatial(interface)
```

What happens when you simulate the same interface, but this time there is a 60 Ohm load?

```python
interface_2 = gen.generate_interface_data(Load_impedance = '30',show_about=False,L_impedance = '985',L_time='0.3',C_impedance='0.88',C_time='0.53')
```

```python
fig,axs = wp.make_time_interconnect_all(interface_2)
wp.make_time_interconnect_all(interface,ax = axs)

for ax_name in axs:
    axs[ax_name].legend(['load','no-load'])
```
